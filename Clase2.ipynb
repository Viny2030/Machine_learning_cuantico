```python
import pennylane as qml
import numpy as np
```

```python
dispositivo = qml.device("default.qubit", wires = 5)
```

```python
def circuito(theta, w):
    if theta > 0:
        qml.RX(theta, wires = w)
    else:
        qml.PauliX(wires = w)
    qml.Hadamard(wires = w)
    return qml.state()
```

```python
nodo = qml.QNode(circuito, dispositivo)
```

```python
nodo(-1, 0)
```

```python
@qml.qnode(dispositivo)
def nodo_directo(theta, w):
    if theta > 0:
        qml.RX(theta, wires = w)
    else:
        qml.PauliX(wires = w)
    qml.Hadamard(wires = w)
    return qml.state()
```

```python
nodo_directo(-1, 0)
```

```python
dispositivo = qml.device("default.qubit", wires = 2)
```

```python
@qml.qnode(dispositivo)
def estado_bell():
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0, 1])
    return qml.state()
```

```python
estado_bell()
```

```python
@qml.qnode(dispositivo)
def probabilidades_bell():
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0, 1])
    return qml.probs()
```

```python
probabilidades_bell()
```

```python
dispositivo_muestras = qml.device("default.qubit", shots = 1024)

@qml.qnode(dispositivo_muestras)
def muestras_bell():
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0, 1])
    return qml.sample()

muestras_bell(shots = 10)
```

```python
medida0 = qml.Hermitian(np.array([[0, 0], [0, 1]]), wires = 0)
```

```python
@qml.qnode(dispositivo)
def valor_esperado_bell():
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0, 1])
    return qml.expval(medida0)

valor_esperado_bell()
```

```python
@qml.qnode(dispositivo)
def circuito_bell_param(theta):
    qml.RX(theta, wires = 0)
    qml.Hadamard(wires = 0)
    qml.CNOT(wires = [0, 1])
    return qml.sample()
```

```python
def circuito(parametros):
    n = len(parametros)
    i = 0
    for p in parametros:
        qml.RX(p, wires = i)
        qml.CNOT(wires = [i, i + 1])
        i += 1
    return qml.probs(wires = [n])
```

```python
dispositivo = qml.device("default.qubit", 8)
param = [2, 5, 1, 4, 7, 0]
nodo = qml.QNode(circuito, dispositivo)
nodo(param)
```
