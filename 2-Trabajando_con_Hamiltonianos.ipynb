##"## Definiendo vectores de estado"

```python
from qiskit.quantum_info import Statevector
zero = Statevector([1,0])
print("zero es un estado de", zero.num_qubits, "qubit")
print("Su dimensión como vector es", zero.dim)
print("Sus amplitudes son", zero.data)
```

```python
one = Statevector([0,1])
print("one es un estado de", one.num_qubits, "qubit")
print("Su dimensión como vector es", one.dim),
print("Sus amplitudes son", one.data)
```

```python
zero = Statevector.from_int(0, dims = 2)
print("zero es un estado de", zero.num_qubits, "qubit")
print("Su dimensión como vector es", zero.dim)
print("Sus amplitudes son", zero.data)
```

```python
one = Statevector.from_int(1, dims = 2)
print("one es un estado de", one.num_qubits, "qubit")
print("Su dimensión como vector es", one.dim)
print("Sus amplitudes son", one.data)
```

```python
psi = one.tensor(zero.tensor(zero))
print("psi es un estado de", psi.num_qubits,"qubits")
print("Su dimensión como vector es", psi.dim)
print("Sus amplitudes son", psi.data)
```

```python
psi = one^zero^zero
psi.draw("latex")
```

```python
psi = Statevector.from_int(4, dims = 8)
psi.draw("latex")
```

```python
from numpy import sqrt
ghz = 1/sqrt(2)*(zero^zero^zero) + 1/sqrt(2)*(one^one^one)
ghz.draw("latex")
```

#### Definiendo hamiltonianos"

```python
from qiskit.quantum_info import Pauli
Z0Z1 = Pauli("ZZI")
print("Z0Z1 es",Z0Z1)
print("Y su matriz es:")
print(Z0Z1.to_matrix())
```

```python
print("La representación dispersa de Z0Z1 es")
print(Z0Z1.to_matrix(sparse=True))
```

```python
Z0Z1 = Pauli(([0,1,1],[0,0,0]))
print("Z0Z1 es",Z0Z1)
```

```python
from qiskit.quantum_info import SparsePauliOp
H_cut = SparsePauliOp(Pauli("ZZI")) + SparsePauliOp(Pauli("ZIZ"))
print("H_cut es", H_cut.to_list())
print("Su matriz es:")
print(H_cut.to_matrix())
print("Su representación dispersa es:")
print(H_cut.to_matrix(sparse=True))
```

```python
I = SparsePauliOp(Pauli("I"))
Z = SparsePauliOp(Pauli("Z"))
H_cut = (Z^Z^I) + (Z^I^Z)
print("H_cut es", H_cut.to_list())
```

```python
H_ising = -0.5*(Z^Z^I) + 2*(Z^I^Z) -(I^Z^Z) + (I^Z^I) -5*(I^I^Z)
print("H_ising es", H_ising.to_list())
```

#### Calculando valores esperados"

```python
print("El valor esperado de psi en H_cut es", psi.expectation_value(H_cut))
```

```python
print("El valor esperado de psi en H_cut es", psi.inner(psi.evolve(H_cut)))
```
