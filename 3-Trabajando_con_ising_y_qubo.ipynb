```python
import dimod
# Particionar (2,5,4,7)\n",
J = {(0,1):20, (0,2):16, (0,3):28, (1,2):40, (1,3):70, (2,3):56}
h = {}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)
print("Nuestro problema es:")
print(problem)
```

```python
solver = dimod.ExactSolver()
solution = solver.sample(problem)
print("Las posibles soluciones son")
print(solution)
```

```python
print("Una de las soluciones óptimas es", solution.first.sample)
print("Tiene energía", solution.first.energy)
```

## SUBSET SUM"

```python
# Encontrar un subconjunto de {1, 4, -2} cuya suma sea lo más cercana posible a 2\n",
J = {(0,1):8, (0,2):-4, (1,2):-16}
h = {0: -3, 2: 12}
problem = dimod.BinaryQuadraticModel(h, J, 4.0, dimod.BINARY)
print("Nuestro problema es:")
print(problem)
```

```python
solver = dimod.ExactSolver()
solution = solver.sample(problem)
print("Las posibles soluciones son")
print(solution)
```

```python
print("La solución óptima es", solution.first.sample)
print("Tiene energía", solution.first.energy)
```

## Conviertiendo QUBO en Ising"

```python
J = {(0,1):-4}
h = {0: 3, 1:4}
qubo = dimod.BinaryQuadraticModel(h, J, 0, dimod.BINARY)
print("Nuestro problema es:")
print(qubo)
```

```python
ising = qubo.to_ising()
print("El problema transformado es:")
print(ising)
```
