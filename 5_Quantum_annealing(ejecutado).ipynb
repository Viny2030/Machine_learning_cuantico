## Resolviendo un problema sencillo (Ising)

```python
BinaryQuadraticModel({0: 0.0, 1: 0.0, 2: 0.0}, {(1, 0): 1.0, (2, 0): 1.0}, 0.0, 'SPIN')
```

```python
import dimod

# El objeto BQM debe ser llamado como dimod.BinaryQuadraticModel
problem = dimod.BinaryQuadraticModel(
    {0: 0.0, 1: 0.0, 2: 0.0},
    {(1, 0): 1.0, (2, 0): 1.0},
    0.0,
    'SPIN'
)

print(problem)
```

```python
import dimod
J = {(0,1):1, (0,2):1}
h = {}
problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)
print("Nuestro problema es:")
print(problem)
```

```python
from dwave.system import DWaveSampler
from dwave.system import EmbeddingComposite
sampler = EmbeddingComposite(DWaveSampler())
result = sampler.sample(problem, num_reads=30)
print("Las soluciones obtenidas son:")
print(result)
```

## Forzando el sistema"


```python
def grafo_circular(N):\n",
        "  J = {}\n",
        "  for i in range(N-1):\n",
        "    J[(i,i+1)] = 1\n",
        "  J[(N-1,0)] = 1\n",
        "  h = {}\n",
        "  problem = dimod.BinaryQuadraticModel(h, J, 0.0, dimod.SPIN)\n",
        "  return problem"
```
